```python
#Graphs and tables or statistics
df['number_of_siblings']=df['H4WS1']
```

```python
df['number_of_siblings'].dtype
```

```python
df['number_of_siblings']= pd.to_numeric(df['number_of_siblings'],errors= 'coerce')
df['number_of_siblings'].dtype
```

```python
df['number_of_siblings'] = df['number_of_siblings'].loc[df['number_of_siblings'] !=96] 
df['number_of_siblings'] = df['number_of_siblings'].loc[df['number_of_siblings'] !=98]
df['number_of_siblings'].value_counts() 
```

```python
import seaborn as sns
```

```python
sns.histplot(x= 'number_of_siblings', data = df, bins = 15)
```

```python
sns.boxplot(x= 'number_of_siblings', data = df)
```

```python
sns.histplot( x= 'number_of_siblings', data=df, bins = 15, color ='green', kde= True)
```

```python
df['high_school_status']=df['H4ED1']
```

```python
df['high_school_status'].dtype
```

```python
df['high_school_status'] = df['H4ED1']
df['high_school_status'] = df['high_school_status'].loc[ df['high_school_status'] !=' ' ]
df['high_school_status'] = df['high_school_status'].loc[ df['high_school_status'] !='6' ]
df['high_school_status'] = df['high_school_status'].loc[ df['high_school_status'] !='8' ]
df['high_school_status'].value_counts()
```

```python
df['high_school_status'].replace(['1','2','3','4'], ['diploma','diploma','diploma','no diploma'],inplace=True)
df['high_school_status'].value_counts()
```

```python
df['high_school_status'].value_counts(normalize=True).plot( kind = 'bar')
```

```python
df['high_school_status'].value_counts(normalize=True).plot( kind = 'pie')
```

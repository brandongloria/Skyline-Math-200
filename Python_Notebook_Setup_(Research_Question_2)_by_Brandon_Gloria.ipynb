Research Question #2: Is there an association between high school graduation status and born as a US Citizen?

Screenshot 2023-05-03 at 10.20.33 AM.png

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

```python
df['high_school_status']=df['H4ED1']
```

```python
df['high_school_status'].dtype
```

```python
df['high_school_status'] = df['H4ED1']
df['high_school_status'] = df['high_school_status'].loc[ df['high_school_status'] !=' ' ]
df['high_school_status'] = df['high_school_status'].loc[ df['high_school_status'] !='6' ]
df['high_school_status'] = df['high_school_status'].loc[ df['high_school_status'] !='8' ]
df['high_school_status'].value_counts()
```

```python
df['high_school_status'].replace(['1','2','3','4'], ['diploma','diploma','diploma','no diploma'],inplace=True)
df['high_school_status'].value_counts()
```

```python
df['high_school_status'].value_counts(normalize=True).plot( kind = 'bar')
```

```python
df['high_school_status'].value_counts(normalize=True).plot( kind = 'pie')
```

```python
df['born_us_citizen']=df['H4OD4']
```

```python
df['born_us_citizen'].dtype
```

```python
df['born_us_citizen']=df['H4OD4']
df['born_us_citizen'] = df['born_us_citizen'].loc[ df['born_us_citizen'] !='6' ]
df['born_us_citizen'] = df['born_us_citizen'].loc[ df['born_us_citizen'] !=' ' ]
df['born_us_citizen'].value_counts()
```

```python
df['born_us_citizen'].replace(['0','1'], ['no','yes'],inplace=True)
df['born_us_citizen'].value_counts()
```

```python
df['born_us_citizen'].value_counts(normalize=True).plot( kind = 'bar')
```

```python
df['born_us_citizen'].value_counts(normalize=True).plot( kind = 'pie')
```
